```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from xgboost import XGBClassifier
```

```python
data = pd.read_csv('testtrainvalidatedata.csv')
excluded_nba_goats = pd.read_csv('excluded_nba_goats.csv')
```

```python
data.info()
```

```python
drop_columns = ["Player","HOF","Pos"]
data.drop(labels = drop_columns, axis = 1, inplace= True)
```

```python


data.fillna(value=0.0, inplace=True) #we will replace NaN values with 0
```

```python
##excluded_nba_goats
excluded_nba_goats
```

```python
Y = data['VALUE']
data.drop(labels = "VALUE", axis = 1, inplace = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, Y, test_size=0.2, random_state=1)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
```

```python
print(len(X_train), len(X_val), len(X_test))
```

```python
# default but can add params if i want
xgb_clf = XGBClassifier()

# fit the model with the training data
xgb_clf.fit(X_train, y_train)
```

```python
# predict the target on the validation dataset
predictions = xgb_clf.predict(X_val)
print(predictions)
```

```python
# we can evaluate the classifier by checking its accuracy and creating a confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_val, predictions))

print("Classification Report")
print(classification_report(y_val, predictions))
```

```python
# predict on the test dataset
prediction_test = xgb_clf.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(y_test, prediction_test))

print("Classification Report")
print(classification_report(y_test, prediction_test))
```
